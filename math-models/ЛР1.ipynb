# Лабораторная работа №1. Методы одномерной оптимизации

Цель работы: изучение и анализ поисковых алгоритмов минимизации функции одной переменной: дихотомического, Фибоначчи и «золотого сечения».

```python
import math
```

```python
a = 0
b = 3

def f(x):
  return -x / math.exp(x)
```

функция нахождения чисел Фибоначчи

```python
fib_val = {0: 0, 1: 1}

def fib(x):
  if x not in fib_val:
    fib_val[x] = fib(x-1) + fib(x-2)
  return fib_val[x]
```

пример

```python
fib(10)
```

значение функции в точке минимума

$$f(x) = \frac{-x}{e^x}$$

$$f'(x) = \frac{x-1}{e^x} = 0$$

$$f'(1) = 0$$

```python
f(1)
```

Метод дихотомии

```python
def dichotomy(a, b, eps=1e-3):
  iters = 0
  while b-a > 2*eps:
    iters += 1
    x = (a+b)/2
    x1 = x - eps/2
    x2 = x + eps/2
    f1 = f(x1)
    f2 = f(x2)

    if (f1 > f2): a = x1
    else: b = x2
  
  return dict(value=f((a+b)/2), iters=iters)
```

```python
dichotomy(a, b)
```

Метод Фибоначчи

```python
def fibonacci(a, b, n):
  L = b - a
  iters = 0
  while (n > 2):
    iters += 1
    x1 = a + fib(n-1)/fib(n) * L
    x2 = b - fib(n-1)/fib(n) * L

    f1 = f(x1)
    f2 = f(x2)

    if (f1 > f2):
      b = x1
      f1 = f2
      x1 = x2
      L = b-a

      x2 = a + (b - x1)
      f2 = f(x2)

    else:
      a = x2
      f2 = f1
      x2 = x1
      L = b-a

      x1 = b - (x2 - a)
      f1 = f(x1)
    
    n -= 1
  
  return dict(value=min(f1, f2), iters=iters)
```

```python
fibonacci(a, b, 10)
```

Метод золотого сечения

```python
def gss(a, b, eps=1e-3):
  L = b-a
  iters = 0
  while L > eps:
    iters += 1
    x1 = a + L * 0.618
    x2 = b - L * 0.618

    f1 = f(x1)
    f2 = f(x2)

    if f1 > f2:
      b = x1
      f1 = f2
      x1 = x2
      L = b-a

      x2 = b - L * 0.618
      f2 = f(x2)
    
    else:
      a = x2
      f2 = f1
      x2 = x1
      L = b-a

      x1 = a + L * 0.618
      f1 = f(x1)
    
  return dict(value=min(f1, f2), iters=iters)
```

```python
gss(a, b)
```

Анализ полученных результатов

```python
fmin = f(1) # функция в т.min
f1 = dichotomy(a, b) # дихотомия
f2 = fibonacci(a, b, 20) # фибоначчи
f3 = gss(a, b) # золотое сечение
```

```python
print('точность -', abs(f1['value'] - fmin), ', число итераций -', f1['iters'])
print('точность -', abs(f2['value'] - fmin), ', число итераций -', f2['iters'])
print('точность -', abs(f3['value'] - fmin), ', число итераций -', f3['iters'])
```
