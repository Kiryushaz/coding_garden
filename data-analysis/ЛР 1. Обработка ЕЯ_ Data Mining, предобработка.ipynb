```python
from urllib.request import urlopen
import json

def download_json(url):
  response = urlopen(url)
  data_json = json.loads(response.read())
  return data_json
  
#download_json('https://api.hh.ru/vacancies/43551857')
```

```python
'''
РЕАЛИЗОВАТЬ remove_garbage
Используйте регулярные выражения
IN: str '<p>ТРЕБОВАНИЯ: <br /> - обязательно высшее или незаконченное высшее экономическое образование;...'
OUT: str 'ТРЕБОВАНИЯ: - обязательно высшее или незаконченное высшее экономическое образование;...'
'''
def remove_garbage(raw_text):
  import re
  text = re.sub(re.compile('<.*?>|&.*?;'), '', raw_text)
  return text
```

```python
clean_desc = remove_garbage(description)
clean_desc
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
sent_tokenize(clean_desc)
```

```python
'''
РЕАЛИЗОВАТЬ tokenize
Используйте word_tokenize из nltk.tokenize
IN: str 'ТРЕБОВАНИЯ: - обязательно высшее или незаконченное высшее экономическое образование;...'
OUT: list ['требования', ':', '-', 'обязательно', 'высшее', 'или', 'незаконченное', 'высшее', 'экономическое', 'образование']
'''
def tokenize(raw_text):
  tokens = word_tokenize(raw_text)
  return tokens
```

```python
tokens = tokenize(clean_desc.lower())
tokens
```

```python
'''
РЕАЛИЗОВАТЬ to_base_form
Можно использовать и лемматизацию и стемминг: pymystem3, nltk, Spacy, gensim, pymorphy 
IN: str 'Красивая мама красиво мыла раму'
OUT: str 'красивый мама красиво мыть рама'
'''
import spacy
nlp = spacy.load('ru_core_news_md')

def to_base_form(raw_text):
  tokens = [text.lemma_ for text in nlp(raw_text)]
  return tokens
```

```python
base_form_tokens = to_base_form(clean_desc)
```

```python
'''
РЕАЛИЗОВАТЬ remove_stop_words
Используйте stopwords из nltk.corpus 
IN: list ['требования', ':', '-', 'обязательно', 'высший', 'или', 'незаконченный', 'высший', 'экономический', 'образование']
OUT: list ['требование', 'обязательно', 'высший', 'незаконченный', 'высший', 'экономический', 'образование']
'''
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(raw_text):
  text = [text for text in raw_text if text.isalpha() and text not in stopwords.words('russian')]
  return text
```

```python
clean_tokens = remove_stop_words(base_form_tokens)
print(tokens)
print(base_form_tokens)
print(clean_tokens)
```

```python
'''
РЕАЛИЗОВАТЬ get_prepared_vacancy_description_from_hh
Используйте реализованные ранее методы
IN: int 72323
OUT1: str vacancy_json dump
OUT2: list[list] [['требование', 'обязательно', 'высший', 'незаконченный', 'высший', 'экономический', 'образование', ...], [...], ...]
'''
def get_prepared_vacancy_from_hh(vacancy_id):
  base_url = 'https://api.hh.ru/vacancies/'
  vacancy_json = download_json(base_url + str(vacancy_id))
  vacancy_description = vacancy_json['description']
  vacancy_description_sentences = sent_tokenize(vacancy_description)
  prepared_vacancy_description = []

  for sentence in vacancy_description_sentences:
    clean_sentence = remove_garbage(sentence)
    tokens = tokenize(clean_sentence.lower())
    base_form_tokens = to_base_form(clean_sentence)
    # (to_base_form and tokenize) or (tokenize and to_base_form)
    prepared_vacancy_description.append(remove_stop_words(base_form_tokens))
  return vacancy_json, prepared_vacancy_description
```

```python
'''
РЕАЛИЗОВАТЬ get_prepared_all_vacancies_from_hh
Используйте get_prepared_vacancy_from_hh и ThreadPoolExecutor из concurrent.futures
IN: list [int]
OUT: list [vacancy_id, raw_description, prepared_description]
'''
import concurrent.futures

def get_prepared_all_vacancies_from_hh(vacancy_ids):
    result = []

    with concurrent.futures.ThreadPoolExecutor() as executor:
        future_to_url = {executor.submit(get_prepared_vacancy_from_hh, vacancy_id): vacancy_id for vacancy_id in vacancy_ids}
        for future in concurrent.futures.as_completed(future_to_url):
          try:
            vacancy_id = future_to_url[future]
            data = future.result()
            result.append([vacancy_id, *data])
          except Exception as exc:
            print(f'{vacancy_id} сгенерировано исключение: {exc}')
    
    return result
```

```python
with open("vacancy_ids.txt") as f:
  vacancy_ids = f.read().splitlines()
```

```python
step = 10
```

```python
%%time
result = get_prepared_all_vacancies_from_hh(vacancy_ids[2000*(step-1):2000*step])
```

```python
print("success", len(result))
print("failure", 2000-len(result))
```

```python
print("ID:", result[0][0])
print("Raw JSON:", result[0][1])
print("Prepared description:", result[0][2])
```

```python
import csv

header = ['vacancy_id', 'raw_json', 'prepared_description']
with open(f'vacancies_{step}.csv', 'w', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for data in result:
      writer.writerow(data)
```
